```python
!pip install pandas

import pandas as pd

# Daftar nama file CSV
csv_files = ["dataset.csv", "dataset1.csv"]

# List untuk menyimpan data dari setiap file
dataframes = []

# Loop untuk membaca setiap file CSV
for file in csv_files:
    df = pd.read_csv(file)  # Membaca file CSV
    dataframes.append(df)   # Menyimpan DataFrame ke dalam list

# Menggabungkan semua DataFrame
merged_df = pd.concat(dataframes, ignore_index=True)

# Menyimpan hasil penggabungan ke file baru
merged_df.to_csv("merged_file.csv", index=False)

print("Semua file berhasil digabungkan ke 'merged_file.csv'")
```

```python
# prompt: cara mendeteksi duplikat pada csv

import pandas as pd

# Load the merged CSV file
merged_df = pd.read_csv("merged_file.csv")

# Check for duplicates based on all columns
# duplicates = merged_df[merged_df.duplicated(keep=False)]

# If you want to check for duplicates based on specific columns, specify them in the subset parameter
# For example, to check for duplicates based on the 'Tweet' and 'Date' columns
duplicates = merged_df[merged_df.duplicated(subset=['full_text'], keep=False)]

if not duplicates.empty:
    print("Duplicate rows found:")
    print(duplicates)

    # Optionally, remove duplicates
    merged_df.drop_duplicates(inplace=True)  # Keep the first occurrence
    merged_df.drop_duplicates(keep=False, inplace=True)  # Remove all duplicates
    print("\nDataFrame after removing duplicates:")
    print(merged_df)

    # Save the updated DataFrame (optional)
    merged_df.to_csv("merged_file_no_duplicates.csv", index=False)
else:
    print("No duplicate rows found.")
```
